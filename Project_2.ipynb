```python
#creating general functions to change the file type


csv_file = "Crime_Data.csv"
json_file = "data.json"
table_name = "sql_table"
connection = sqlite3.connect("sql_table")

def json_to_csv():
  df = pd.DataFrame(weather_data)
  print(f"{df.shape}") #printing details of the data set before cols change
  df = df.drop(df.columns[[-1]], axis=1)
  df.to_csv(csv_file, index=False)
  print(f"JSON data converted to CSV and saved as {csv_file}") #saving the new file
  print(f"{df.shape}") #printing details of the data set after cols change

def json_to_sql():
  df = pd.DataFrame(weather_data)
  print(f"{df.shape}") #printing details of the data set before cols change
  df = df.drop(df.columns[[-1]], axis=1)
  df.to_sql(table_name, connection, if_exists='replace', index=False)
  print(f"JSON data converted to SQL and saved as {table_name}") #saving the new file
  print(f"{df.shape}") #printing details of the data set after cols change

def csv_to_json():
  df = pd.read_csv(csv_file)
  print(f"{df.shape}") #printing details of the data set before cols change
  df = df.drop(df.columns[[-1]], axis=1)
  df.to_json(json_file, orient="records")
  print(f"CSV data converted to JSON and saved to {json_file}") #saving the new file
  print(f"{df.shape}") #printing details of the data set after cols change
```

```python
def main():
  input_format = input("Enter input format (json, csv): ").lower()
  output_format = input("Enter output format (json, csv, sql): ").lower()

  if input_format == "json":
    if output_format == "sql":
      json_to_sql()
    elif output_format == "csv":
      json_to_csv()
  elif input_format == "csv":
    if output_format == "json":
      csv_to_json()
  else:
        print("Invalid File Type")
```

```python
main()


```
