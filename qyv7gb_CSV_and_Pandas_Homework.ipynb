```python
import pandas as pd
```

```python
acc_df = pd.read_csv("acc.csv")
```

```python
#display first 5 rows of the data frame
acc_df.head(5)

```

```python
acc_df.columns = acc_df.iloc[0]
acc_df = acc_df[1:]
```

```python
acc_df.head()
```

```python
#add up the total points scored by all players
#points is stored as a string, have to convert to numeric

acc_df["PTS"] = pd.to_numeric(acc_df["PTS"])


acc_df["MP"] = pd.to_numeric(acc_df["MP"])

```

```python

total_points = acc_df["PTS"].sum()
total_points
```

```python
#find the player who has the most minutes

most_minutes = acc_df[['Player', 'MP']].sort_values(by='MP', ascending=False).head(1)
most_minutes
```

```python
#Identify the top 5 players for total rebounds
acc_df["TRB"] = pd.to_numeric(acc_df["TRB"])

total_rebounds = acc_df[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)
total_rebounds
```

```python
#create a df containing only players that played more than 500 minutes

new_df = acc_df[acc_df["MP"] > 500]
```

```python
new_df.head(5)
```

```python
#with new df, player with highest number of total assists (AST)

new_df["AST"] = pd.to_numeric(new_df["AST"])

total_assists = new_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(1)
total_assists
```

```python
#top 3 assist leads in the league

three_assistsLeads = new_df[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)
three_assistsLeads
```

```python
#top 3 shot blockers
new_df["BLK"] = pd.to_numeric(new_df["BLK"])


shot_blockers = new_df[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)
shot_blockers
```

```python
#Group the players by school and calculate the total points scored by each school

new_df["PTS"] = pd.to_numeric(new_df["PTS"])
```

```python
total_points = new_df.groupby('School')[['PTS']].sum().reset_index()
total_points
```

```python
#groupby school and total assists

new_df["AST"] = pd.to_numeric(new_df["AST"])
total_assists = new_df.groupby('School')[['AST']].sum().reset_index()
```

```python
total_assists
```

```python
#sort total_points to get top 3

total_points.sort_values(by="PTS", ascending = False).head(3)

```

```python
#Extra Credit- create a bar chart showing the top 5 players by total points scored

top_scorers = acc_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)
top_scorers
```

```python
#Plot top 5 using bar chat

import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(10,6))
sns.barplot(x='Player', y='PTS', data=top_scorers)
plt.title('Top 5 ACC scorers')
plt.show()
```

```python
#correlation between FG% and WS

acc_df["FG%"] = pd.to_numeric(acc_df["FG%"])

plt.figure(figsize=(10,6))
plt.scatter(acc_df["FG%"], acc_df["WS"])
plt.title('Correlation between Field Goal Percentage and Win Shares')
plt.xlabel("Field Goal Percentage")
plt.ylabel("Win Shares")
plt.show()
```

```python
#find the correlation

correlation = acc_df["FG%"].corr(acc_df["WS"])
print(f"\nThe correlation is: {correlation}")
```
