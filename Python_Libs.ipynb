```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Crime_Data.csv")
```

```python
data.head()
```

```python
#bar chart

area_counts = data['AREA'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.bar(area_counts.index, area_counts.values)
plt.title('Distribution of AREA')
plt.xlabel('Area')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#line charts

plt.figure(figsize=(10, 6))
plt.plot(data['LON'], data['LAT'])
plt.title('Geographical Distribution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)  # Add a grid for better visualization
plt.tight_layout()
plt.show()
```

```python
#histogram

plt.figure(figsize=(10, 6))
plt.hist(data['Vict Age'], bins=10)  # Adjust the number of bins as needed
plt.title('Distribution of Victim Age')
plt.xlabel('Victim Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#exploration using sns

plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.boxplot(x=data['Vict Age'])
plt.title('Distribution of Victim Age')
plt.xlabel('Victim Age')
plt.tight_layout()
plt.show()
```

```python
#violin plot using seaborn

plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.violinplot(x='Vict Sex', y='Vict Age', data=data)
plt.title('Distribution of Victim Age by Sex')
plt.xlabel('Victim Sex')
plt.ylabel('Victim Age')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
sns.boxplot(x=data['Vict Sex'])
plt.title('Distribution of Victim Sex')
plt.xlabel('Victim Sex')
plt.tight_layout()
plt.show()
```
